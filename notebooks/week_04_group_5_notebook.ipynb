```python
import numpy as np
import pandas as pd
```

```python
all_sheets_dict = pd.read_excel('https://docs.google.com/spreadsheets/d/1ecopK6oyyb4d_7-QLrCr8YlgFrCetHU7-VQfnYej7JY/export?format=xlsx', sheet_name=None)

all_sheets_dict['amsterdam_weekdays']
```

```python
list_of_dfs = []
for sheet_name, df in all_sheets_dict.items():
    df['sheet_name'] = sheet_name
    list_of_dfs.append(df)

new_df = pd.concat(list_of_dfs, ignore_index=True)
new_df
```

```python
new_df['city'] = new_df['sheet_name'].apply(lambda x: x.split('_')[0])

new_df['weekdays'] = new_df['sheet_name'].apply(lambda x: x.split('_')[1])
new_df
```

```python
new_df['city'].unique()
```

```python
city_country_map = {
    'amsterdam': 'Netherlands',
    'athens': 'Greece',
    'berlin': 'Germany',
    'barcelona': 'Spain',
    'budapest': 'Hungary',
    'lisbon': 'Portugal',
    'london': 'United Kingdom',
    'paris': 'France',
    'rome': 'Italy',
    'vienna': 'Austria'
}
new_df['country'] = new_df['city'].map(city_country_map)
new_df
```

```python
new_df = new_df.rename(columns={'realSum': 'Price'})
new_df
```

```python
new_df = new_df.drop(columns=[new_df.columns[0]])
new_df
```

```python
def set_room_category(row):
    if row['room_private'] == True:
        return 'Private room'
    elif row['room_shared'] == True:
        return 'Shared room'
    else:
        return 'Entire home/apt'

new_df['room_type'] = new_df.apply(set_room_category, axis=1)
new_df = new_df.drop(columns=['room_private', 'room_shared'], axis=1)
```

```python
def combine_multi_and_biz(row):
    if row['biz'] == 1:
        return 'more than four'
    elif row['multi'] == 1:
        return 'two to four'
    else:
        return 'one'
    
new_df['host_listing_category'] = new_df.apply(combine_multi_and_biz, axis=1)
new_df = new_df.drop(columns=['multi', 'biz'])
```

```python
# 5. Are ratings typically high across listings, or is there a wide variation? 
# a. Plot the distribution of guest_satisfaction_overall ratings.  
# b. Examine the distribution of cleanliness_rating.

import matplotlib.pyplot as plt
import seaborn as sns   
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(new_df['guest_satisfaction_overall'], bins=20, kde=True)
plt.title('Distribution of Guest Satisfaction Overall Ratings')
plt.xlabel('Guest Satisfaction Overall Rating')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.histplot(new_df['cleanliness_rating'], bins=20, kde=True)
plt.title('Distribution of Cleanliness Ratings')
plt.xlabel('Cleanliness Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 6. How does person_capacity vary across listings? What is the most common capacity of listings?
person_capacity_counts = new_df['person_capacity'].value_counts().sort_index()
person_capacity_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Distribution of Person Capacity Across Listings')
plt.xlabel('Person Capacity')
plt.ylabel('Number of Listings')
plt.show()
```

```python
# 7. Plot the distribution of price (former realSum) for both weekday and weekend offers
# a. Check normality, skewness, modality
# b. If skewed, consider using transformations (e.g. log); can you normalize it? 
weekday_prices = new_df[new_df['weekdays'] == 'weekdays']['Price']
weekend_prices = new_df[new_df['weekdays'] == 'weekends']['Price']
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(weekday_prices, bins=30, kde=True)
plt.title('Distribution of Prices on Weekdays')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.histplot(weekend_prices, bins=30, kde=True)
plt.title('Distribution of Prices on Weekends')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# b. If skewed, consider using transformations (e.g. log); can you normalize it? It is skewed we will use log transformation
log_weekday_prices = np.log1p(weekday_prices)
log_weekend_prices = np.log1p(weekend_prices)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(log_weekday_prices, bins=30, kde=True)
plt.title('Log-Transformed Distribution of Prices on Weekdays')
plt.xlabel('Log(Price + 1)')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
sns.histplot(log_weekend_prices, bins=30, kde=True)
plt.title('Log-Transformed Distribution of Prices on Weekends')
plt.xlabel('Log(Price + 1)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 8. Is there a difference in price per night between two cities? Choose at least three pairs of cities during weekdays.
weekday_data = new_df[new_df['weekdays'] == 'weekdays']
city_pairs = [('amsterdam', 'berlin'), ('paris', 'rome'), ('london', 'barcelona')]
for city1, city2 in city_pairs:
    prices_city1 = weekday_data[weekday_data['city'] == city1]['Price']
    prices_city2 = weekday_data[weekday_data['city'] == city2]['Price']
    plt.figure(figsize=(10, 5))
    sns.kdeplot(prices_city1, label=city1, fill=True)
    sns.kdeplot(prices_city2, label=city2, fill=True)
    plt.title(f'Price Distribution on Weekdays: {city1} vs {city2}')
    plt.xlabel('Price')
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```

```python
# 9. Does the price per night tend to be higher on weekends?
weekend_mean_price = weekend_prices.mean()
weekday_mean_price = weekday_prices.mean()
weekend_mean_price, weekday_mean_price
# Now we will look at the distributions to see the difference visually
plt.figure(figsize=(10, 5))
sns.kdeplot(weekday_prices, label='Weekdays', fill=True)
sns.kdeplot(weekend_prices, label='Weekends', fill=True)
plt.title('Price Distribution: Weekdays vs Weekends')
plt.xlabel('Price')
plt.ylabel('Density')
plt.legend()
plt.show()

# Now we will look at the distributions to see the difference visually on the log values
plt.figure(figsize=(10, 5))
sns.kdeplot(log_weekday_prices, label='Weekdays', fill=True)
sns.kdeplot(log_weekend_prices, label='Weekends', fill=True)
plt.title('Log-Transformed Price Distribution: Weekdays vs Weekends')
plt.xlabel('Log(Price + 1)')
plt.ylabel('Density')
plt.legend()
plt.show()

# Now we can do the T-test to see if the difference is statistically significant
from scipy import stats
t_stat, p_value = stats.ttest_ind(weekend_prices, weekday_prices, equal_var=False)
t_stat, p_value
# If p-value < 0.05, we can conclude that there is a statistically significant difference in prices between weekends and weekdays
# T-stat is 2.8647968502507535, p-value is 0.004174510702298878
# Since p-value < 0.05, we reject the null hypothesis and conclude that there is a statistically significant difference in prices between weekends and weekdays.
```

```python
# 10. Are listings of superhosts more expensive than those of normal hosts?
superhost_prices = new_df[new_df['host_is_superhost'] == True]['Price']
normalhost_prices = new_df[new_df['host_is_superhost'] == False]['Price']
plt.figure(figsize=(10, 5))
sns.kdeplot(superhost_prices, label='Superhosts', fill=True)
sns.kdeplot(normalhost_prices, label='Normal Hosts', fill=True)
plt.title('Price Distribution: Superhosts vs Normal Hosts')
plt.xlabel('Price')
plt.ylabel('Density')
plt.legend()
plt.show()

# Now we look at the log-transformed prices
log_superhost_prices = np.log1p(superhost_prices)
log_normalhost_prices = np.log1p(normalhost_prices)
plt.figure(figsize=(10, 5))
sns.kdeplot(log_superhost_prices, label='Superhosts', fill=True)
sns.kdeplot(log_normalhost_prices, label='Normal Hosts', fill=True)
plt.title('Log-Transformed Price Distribution: Superhosts vs Normal Hosts')
plt.xlabel('Log(Price + 1)')
plt.ylabel('Density')
plt.legend()
plt.show()

# Now we can do the T-test to see if the difference is statistically significant
t_stat, p_value = stats.ttest_ind(superhost_prices, normalhost_prices, equal_var=False)
t_stat, p_value
# If p-value < 0.05, we can conclude that there is a statistically significant difference in prices between superhosts and normal hosts
# T-stat is -10.787709704306543, p-value is 4.519138249883784e-27
# Since p-value < 0.05, we reject the null hypothesis and conclude that there is a statistically significant difference in prices between superhosts and normal hosts.
```

```python
# 11. Are superhosts closer to the city center and metro station than normal hosts?
superhost_distance_city_center = new_df[new_df['host_is_superhost'] == True]['dist']
normalhost_distance_city_center = new_df[new_df['host_is_superhost'] == False]['dist']
plt.figure(figsize=(10, 5))
sns.kdeplot(superhost_distance_city_center, label='Superhosts', fill=True)
sns.kdeplot(normalhost_distance_city_center, label='Normal Hosts', fill=True)
plt.title('Distance to City Center: Superhosts vs Normal Hosts')
plt.xlabel('Distance to City Center (km)')
plt.ylabel('Density')
plt.legend()
plt.show()

superhost_distance_metro = new_df[new_df['host_is_superhost'] == True]['metro_dist']
normalhost_distance_metro = new_df[new_df['host_is_superhost'] == False]['metro_dist']
plt.figure(figsize=(10, 5))
sns.kdeplot(superhost_distance_metro, label='Superhosts', fill=True)
sns.kdeplot(normalhost_distance_metro, label='Normal Hosts', fill=True)
plt.title('Distance to Metro: Superhosts vs Normal Hosts')
plt.xlabel('Distance to Metro (km)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# 12. Are superhosts cleaner than normal hosts?
# We will compare the cleanliness ratings of superhosts and normal hosts
superhost_cleanliness = new_df[new_df['host_is_superhost'] == True]['cleanliness_rating']
normalhost_cleanliness = new_df[new_df['host_is_superhost'] == False]['cleanliness_rating']
plt.figure(figsize=(10, 5))
sns.kdeplot(superhost_cleanliness, label='Superhosts', fill=True)
sns.kdeplot(normalhost_cleanliness, label='Normal Hosts', fill=True)
plt.title('Cleanliness Ratings: Superhosts vs Normal Hosts')
plt.xlabel('Cleanliness Rating')
plt.ylabel('Density')
plt.legend()
plt.show()

# Now we can do the T-test to see if the difference is statistically significant
t_stat, p_value = stats.ttest_ind(superhost_cleanliness, normalhost_cleanliness, equal_var=False)
t_stat, p_value
# If p-value < 0.05, we can conclude that there is a statistically significant difference in cleanliness ratings between superhosts and normal hosts
# T-stat is 92.22126010432703, p-value is 0.0
# Since p-value < 0.05, we reject the null hypothesis and conclude that there is a statistically significant 
# difference in cleanliness ratings between superhosts and normal hosts.

# Now we can plot the boxplots for better visualization
plt.figure(figsize=(10, 5))
sns.boxplot(x='host_is_superhost', y='cleanliness_rating', data=new_df)
plt.title('Cleanliness Ratings: Superhosts vs Normal Hosts')
plt.xlabel('Is Superhost')
plt.ylabel('Cleanliness Rating')
plt.xticks([0, 1], ['Normal Hosts', 'Superhosts'])
plt.show()      
```
